```python
import pandas as pd
import networkx as nx
import numpy as np
from math import inf
```

```python
df = pd.read_csv("/gdrive/MyDrive/SNDA_PS2.csv", index_col=0)
df = df.replace('-', 0).convert_dtypes(convert_integer=1)
vals = df.values.astype(int)
```

```python
G = nx.DiGraph(vals)
nx.draw(G,with_labels=True)
```

```python
nx.degree_centrality(G)
```

```python
nx.closeness_centrality(G)
```

```python
nx.betweenness_centrality(G)
```

```python
nx.eigenvector_centrality(G)
```
